```python
#1

file_path = '/content/drive/MyDrive/cars.xlsx'

data = pd.read_excel(file_path)

data.head()
```

```python
# a)
import matplotlib.pyplot as plt

plt.figure(figsize = (6,6))
plt.scatter(data['cty'],data['hwy'])
plt.title("City miles per galon vs Highway miles per galon")
plt.xlabel('city')
plt.ylabel('highway')
plt.show()
```

```python
# b)

# when cty increases, hwy also increases
```

```python
# 2

import seaborn as sns

cols = ['displ', 'year', 'cyl', 'cty', 'hwy']
df = data[cols + ['drv']]

sns.pairplot(df, hue='drv', palette='Set2')
plt.suptitle('Scatterplots and Distributions by Drive Type (drv)', y=1.02)
plt.show()
```

```python
#3
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='class', palette='Set2')

plt.title('Count of Car Types (class)')
plt.xlabel('Car Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


#suv
```

```python
#4
plt.figure(figsize=(8, 6))

sns.barplot(data=data, x='drv', y='cyl')

plt.title('Number of Cars per Drive Train (based on Cylinder Count)')
plt.xlabel('Drive Train')
plt.ylabel('Number of Cars')
plt.show()

# r is the most frequent
```

```python
#5
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='class', y='displ', palette='Set2')

plt.title('Engine Displacement by Car Class (Boxplot)')
plt.xlabel('Car Class')
plt.ylabel('Engine Displacement (liters)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x='class', y='displ', palette='Set2')

plt.title('Engine Displacement by Car Class (Violinplot)')
plt.xlabel('Car Class')
plt.ylabel('Engine Displacement (liters)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
